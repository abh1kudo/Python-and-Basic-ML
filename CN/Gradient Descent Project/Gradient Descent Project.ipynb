```python
import numpy as np
import pandas as pd

df = pd.read_csv("C:\Abhinav\Course\CN\Gradient Descent Project/train.csv",delimiter = ',')
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]

#print (x)

#print(x.columns)
#print(df.head())
print(x.shape)
print(y.shape)
```

```python
def addfeatures(x):
    colname = x.columns
    for s in colname:
        x[s + " square"] = x[s]**2
    #x["crimcube"]  = x[colname[0]]**3
    x["scube"] = x[colname[5]]**3
    A = [2,4,5,12,6]
    for i in range(len(A)):
        for j in range(i+1,len(A)):
            x[colname[A[i]] + "into" + colname[A[j]]] = x[colname[A[i]]]*x[colname[A[j]]]
    #adding coefficient wala column
    x["constants"] = np.ones(len(x[colname[0]]))
    return x
```

```python
x1 = addfeatures(x)
print(x1.shape)
```

```python
def cost(X, Y, m) :
    noofrows = len(X.values)
    noofcol = len(X.values[0])
    
    total_cost = 0
    
    for i in range(noofrows) :
        predict = 0
        
        for j in range(noofcol) :
            predict += m[j]*X.values[i, j]
        
        total_cost += ((Y.values[i] - predict)**2)
        
    total_cost /= noofrows
    
    return total_cost
```

```python
def pred(X,M):
    return np.dot(X,M.reshape(len(M),1)).reshape(-1)
    
```

```python
def gradient_descent(lr, epochs , X, Y) :
    N = len(X.values[0])
    M = len(X.values)
    m = np.ones(N)
    for e in range(epochs) :
        m_copy = np.ones(N)
        print(e, ' ', cost(X, Y, m))
        for i in range(N) :
            m_copy[i] = m[i]
        for k in range(N) :
            term = 0
            for i in range(M) :
                y_pred = 0
                for j in range(N) :
                    y_pred += m_copy[j]*X.values[i, j]
                term -= 2*(Y.values[i] - y_pred)*X.values[i, k]
            term /= M
            m[k] = m[k] - (lr*term)
    return m
```

```python
m = gradient_descent(0.01, 200 , x1 , y)

x_test = pd.DataFrame(np.loadtxt('test.csv', delimiter = ','))
x_test = addfeatures(x_test)
#x_test = addBiasColumn(x_test)
#x_test = addPolynomialFeatures(x_test, imp)
y_predict = pred(x_test, m)

np.savetxt('sol.csv', y_predict, fmt = '%.5f')
```
