```python
#datatype is always same in numpy arrays hence less memory is taken also since numpy is itself coded in c , so it is faster
#create numpy arrays
import numpy as np
import math
l = [1,2,3]
print(np.array(l))
l1 = [1,2,"3"]
print(np.array(l1), type(np.array(l1)[0]))

a =  np.zeros(10)
print(a, type(a[0]), a.dtype)
print(a.shape)
b =np.ones(10,int)
print(b, type(b[0]))

#2d arrays , give tuple as argument or soforth for multidimensional arrays

c =  np.zeros((3,4),int)
c[2][2] ="1" #convert string 1 to int 1 
print(c)

#slicing is also allowed 
print(a[2:6])
#slicing in 2d
print(c[:,2]) #gives entire column of index 2

#To know the shape / dimensions 
print(c.shape)

print(np.arange(10))
```

```python
#Operations
#in python lists , we cant add integer to list , but in np arrays , we add the scalar(int) to each element of array
m = [1,2,3]
k = [2,3,4]
l = np.array([1,2,3])
n = np.array([2,3,4])
#print(m+1)  gives error for python list
print(l+1)
print(l+n) #gives element wise addition
print(m+k) #gives append wala kaam
print(2*m) #prints lsit twice
print(2*l) #multiplies each element by 2
print(l**2)
print(l**n)
#print(m**2) gives error 

#mean and sum
print(n.mean())
print(n.sum())

#Euclidean distance in a 3d space
a = np.array([4,5,2])
b = np.array([1,2,9])
d = (a-b)**2
print(d)
dist = math.sqrt(d.sum())
print(dist)

#standard deviation
x = np.array([1,2,3,4,5])
x = (x - x.mean())**2
print(x.sum())

#dot product
a1 = np.array([[1,2],[1,2]])
a2 = np.array([[2,1],[2,1]])
print(a1.dot(a2))

a = np.array([0, 1, 2, 3])
b = np.array([2, 3, 4, 5])
print(a.dot(b))

#matrices useful for dieect matrix products extra , given as np.mat()
```

```python
#pandas
import pandas as pd
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
print(type(iris))
#first row is taken by pandas as column headers
```

```python
#make a copy
ds = iris #shallow copy , only gives reference

#gives the deep copy of the dataframe
df = iris.copy()

#to see initial few entries of the data
print(df.head(5)) #argument is optional to get initial number of rows

#similarly we have df.tail() as well
```

```python
#get the current column headers
print(df.columns)

#to correct the column headers
df.columns = ["sl","sw","pl","pw","flower_type"]
print(df.columns)
```

```python
#to see the shape of the data
print(df.shape)

#to see the datatypes of the data - for each column
print(df.dtypes)
```

```python
#One of the most useful functions to describe the data -  tells us how many nan's are there , mean etc will be found
print(df.describe())
#count tells us how many entries are definite (not nan)
#mean, standard deviation , minimum value , maximum value , spread of value (percentile values - 25 , 50 , 75)
```

```python
#to access a particular column , we can directly access as - df.columnname or df["columnname"]
print(df.sl)
print(df["sl"])
```

```python
#is null tells us if an entry is null
print(df.isnull())
```

```python
#since checking each entry for null is difficult , do sum that gives for each column number of entries that are null
print(df.isnull().sum())
```

```python
#to access data slice (parts of data from the daat frame)
print(df.iloc[1:4 , 2:4]) #works on the basis of position not label
print(df.loc[2]) #wprks pm the basis of label not position
#the above gives rows 1,2,3 [slice is 1:4] and columns 2,3 [slice is 2:4]
```

```python
#removing rows from the data we have drop function (only drops row with label as argument)(labels dont get updated once the dataframe has been loaded)
a = df.drop(1)
#this doesnt do in place removal , makes a copy of the data and removes the argument row with label from the copy
print(df.head(2),"\n",a.head(2))

#to make it inplace , we have 2nd argument as well 
df.drop(0,inplace = True)

print(df.head(2))
```

```python
#to find what all labels/indices are present in the dataframe
print(df.index)
```

```python
print(df.index[0])
#to find the label of first row present in the dataframe
#hence to remove first rows in terms of position
df.drop(df.index[[0,1,2]])
```

```python
#relational operators
print(df.sl > 5) # based on values gives row labels and true or false

```

```python
#also to only get rows which satisfy a particular condition , we can use 
print(df[df.sl>6])
```

```python
#to add a row
df.loc[0] = [1,2,3,4,"Iris-setosa"]
print(df.tail(2))
```

```python
#since our indexes/labels are very weird because of deletion etc, we can reset them
print(df.index)
ab = df.copy()
ab.reset_index()
#the reset_index goves us the indices reset(labels) but adds the old indices into a new column
#also this is not inplace , we have to do (inplace = True)
#if we dont want that to happen , add argument(drop = True)
```

```python
#to delete columns 
ab.drop("sl", axis = 1 , inplace = True)
#axis = 1 makes it a column deletion , inplace for inplace deletion , and column name is needed for deletion
print(ab.head())

#2nd way of deletion is 
del ab["sw"]
print(ab.head())
```

```python
#to add a column that is like difference of two rows ,sum , multiply etc or a constant df["col"] = 1
ab = df.copy()
ab["diff"] = ab["sl"] - ab["pw"]
print(ab.head())
```

```python
#handling NAN
import numpy as np
df.iloc[2:4,1:3] = np.nan   #first make it nan since no entry in the given dataset is nan
```

```python
print(df.describe())
#the count of columns 1 and 2 have gone down since we inserted NAN in there

```

```python
#one way to handle is to remove the rows that have nan values
df.dropna()
#this has removed the rows 2,3 (labels 3,4) which had nan values (again inplace = True is needed for change to be made in df)
```

```python
#other option is to fill with mean , mode etc
df.sw.fillna(df.sw.mean())
#rows with label 3,4 have been filled with mean (inplace = True is needed for the inplace change)
```

```python
#handling strings in data
df["Gender"] = "Female"
print(df.head())
df.iloc[1:3,5] = "Male"
print(df.head())
```
