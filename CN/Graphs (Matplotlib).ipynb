```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [3,4,5]

plt.scatter(x,y)
plt.show()
```

```python
#to connect the points with a line
plt.plot(x,y)
plt.show()
```

```python
#to show the points as well on the line
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
#we can define colors for the graph as 3rd argument as string , 1st character is for color , 2nd is for shape
plt.plot(x,y,"r-.")
plt.show() # argument color = "red" also works
#also marker = "p" also works as argument
#r- gives line , r-- gives dashed line , ro gives points , r+ gives points in the shape of + , rp gives pentagons , r-. gives line of these forms
```

```python
#plot function x^3
import numpy as np
x = np.array([1,2,3,4])
y = x**3
plt.plot(x,y)
plt.show()
```

```python
#get smoother curves
x =  np.arange(0,5,0.1)
y = x**3
plt.plot(x,y,"bo")
plt.show()
```

```python
a =[3,4,5]
plt.plot(a , linewidth = 5) # line width as argument also works
plt.show() #plotting a single array takes values of the array as y , with x as index
```

```python
#Now we want to give labels to axes and title
#get smoother curves
x =  np.arange(0,5,0.1)
y = x**3
plt.plot(x,y,"bo")
plt.xlabel("Values of X")
plt.ylabel("Values of X^3")
plt.title("Cubical Graph")
plt.show()
```

```python
#label or legend for the graph
x =  np.arange(0,5,0.1)
y = x**3
y2 = x
plt.plot(x,y, label = "x^3")
plt.plot(x,y2,label = "Sline")
plt.xlabel("Values of X")
plt.ylabel("Y")
plt.legend()
plt.title("Graph")
plt.show()
```

```python
#change size of axes

x =  np.arange(0,5,0.1)
y = x**3
y2 = x
plt.plot(x,y, label = "x^3")
plt.plot(x,y2,label = "Sline")
plt.xlabel("Values of X")
plt.ylabel("Y")
plt.legend()
#it asks for a list as argument, with 4 entries , start and end of x and y respectively
plt.axis([1,10,1,45])

plt.grid() #to make a grid


#to add text at a particular place
plt.text(6,20, "testing text", fontsize=12) #starting point for the text

plt.title("Graph")
plt.show()
```

```python
#Pie graphs
sizes = [1,2,3,4]
labels = ["A","B","C","D"]
colors = ["blue","red", "green","purple"]
plt.pie(sizes,colors=colors,labels = labels,autopct="%.0f%%") #take array of sizes , double % sign for showing % on graph
plt.axis("equal") # to make it a perfect circle

#to show the percentage values of each region
#autopct argument is needed , which takes a function, which returns what string to show
plt.show()
```

```python
#explode property to remove a particular part from the graph and show separately
sizes = [1,2,3,4]
labels = ["A","B","C","D"]
colors = ["blue","red", "green","purple"]
explodes =  [0.2,0,0.5,0]
plt.pie(sizes,colors=colors,labels = labels,autopct="%.0f%%", explode=explodes, counterclock=False,startangle=90) 
plt.axis("equal") 
plt.show()
#default graph is counterclockwise
#default start angle is x axis
```
