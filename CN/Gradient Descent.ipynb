### Humara Gradient Descent will have 4 Arguments
1- X <br>
2- Y <br>
3- Learning Rate <br>
4- Number of Iterations <br>

```python
def stepgd(x,y,lr,m,c):
    noofrows = len(x)
    derm = 0 #this will be change in m (sigma ke liye for loop lagaya hai)
    derc = 0 #this will be change in c
    for i in range(noofrows):
        derm+= (-2/noofrows)*(y[i]-m*x[i]-c)*(x[i])
        derc+= (-2/noofrows)*(y[i]-m*x[i]-c)
    m = m - lr*derm
    c = c - lr*derc
    return m,c
```

There are 3 types of Gradient descents - <br>
1- __Batch Gradient Descent__ , jiska code upar likha hai , and jisme derm poora calculate hoga for all rows in 1 iteration
   and finally m change hoga at the end of an iteration <br>
2- __Stochastic Gradient Descent__ - Jisme we keep on updating m after every row se calculating derivative value , toh isme     saare derm sum nhi hote, balki m update hota hai <br>
3- __Mini Batch Gradient Descent__ - Isme mixture of both hai , kuch batch pehle derm banata hai (not all rows) , fir m update hota hai uss value se

```python
def cost(x,y,m,c):
    ans = 0
    noofrows = len(x)
    for i in range(noofrows):
        ans+= (1/noofrows)*((y[i]-m*x[i]-c)**2)
    return ans
```

```python
def gd(x,y,lr,noi):
    m,c = 0,0 #random value se initialise karo
    for i in range(noi):
        #step wala gradient kaam karo
        m,c = stepgd(x,y,lr,m,c)
        print(i+1 ,"Iteration Ka Cost = " ,cost(x,y,m,c))
    return m,c
```

__Now We will do for Generic Gradient Descent , ie more than 1 variable (or n feature data) , isme bass c hatt jaega and m will not be an integer but an array of size n+1 , (n features and 1 added for adding c to the array)__

```python
def stepggd(x,y,m,lr):
    noofrows = len(x)
    derm = np.zeros(len(x[0]))
    for i in range(len(x)):
        error = (m*x[i]).sum()
        error = y - error
        for j in range(len(x[0])):
            derm[j] += (-2/noofrows)*(error)*(x[i][j])
    for i in range(len(x[0])):
        m[i] = m[i] - lr*derm[i]
    return m
```

```python
def ggd(x,y,lr,noi):
    m = np.zeros(len(x[0]))
    for i in range(noi):
        m = stepggd(x,y,m,lr)
    return m
```
